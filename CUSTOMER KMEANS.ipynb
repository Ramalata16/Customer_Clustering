```python
# FEATURE ENGINEERING

df["PURCHASE_TO_LIMIT"] = (
    df["PURCHASES"] /
    (df["CREDIT_LIMIT"] + 1)
)

df["PAYMENT_RATIO"] = (
    df["PAYMENTS"] /
    (df["MINIMUM_PAYMENTS"] + 1)
)

df["CASH_ADVANCE_RATIO"] = (
    df["CASH_ADVANCE"] /
    (df["PURCHASES"] + 1)
)

df["AVG_PURCHASE_VALUE"] = (
    df["PURCHASES"] /
    (df["PURCHASES_TRX"] + 1)
)

df["AVG_CASH_ADVANCE"] = (
    df["CASH_ADVANCE"] /
    (df["CASH_ADVANCE_TRX"] + 1)
)


cols = [
    "BALANCE",
    "PURCHASES",
    "CASH_ADVANCE",
    "PAYMENTS",
    "MINIMUM_PAYMENTS"
]

for col in cols:
    df[col] = np.log1p(df[col])
```

```python
# SCALING


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)

X_pca = pca.fit_transform(X_scaled)

print(X_pca.shape)
```

```python
# IMPORTING KMEANS TO CREATE DIFFERENT CLUSTERS

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2,11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(X_pca)

    score = silhouette_score(X_pca, labels)

    print(f"k={k}: {score:.3f}")
```

```python
from sklearn.cluster import KMeans

for k in [2,3,4]:
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    df[f'Cluster_{k}'] = kmeans.fit_predict(X_pca)
```

```python
cluster_profile = df.groupby('Cluster_4').mean()

cluster_profile
```

```python
cluster_names = {
    0: "Installment Shoppers",
    1: "High-Risk Credit Users",
    2: "Cash Advance Dependents",
    3: "Premium Active Customers"
}

df["Customer_Segment"] = df["Cluster_4"].map(cluster_names)

from sklearn.decomposition import PCA
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
pca_2d = pca.fit_transform(X_scaled)

plot_df = pd.DataFrame(
    pca_2d,
    columns=["PC1", "PC2"]
)

plot_df["Customer_Segment"] = df["Customer_Segment"]

plt.figure(figsize=(12,8))

sns.scatterplot(
    data=plot_df,
    x="PC1",
    y="PC2",
    hue="Customer_Segment",
    s=60
)

plt.title("Customer Segmentation Using K-Means")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Customer Segment")
plt.show()
```
